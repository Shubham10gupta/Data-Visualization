```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
print(iris_df.head())
```

```python
sns.set(style="whitegrid")
```

```python
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.boxplot(data=iris_df, x='species', y='sepal length (cm)')
plt.title("Boxplot of Sepal Length by Species")
plt.show()
```

```python
sns.boxplot(data=iris_df, x='species', y='sepal width (cm)')
plt.title("Boxplot of Sepal Width by Species")
plt.show()
```

```python
sns.boxplot(data=iris_df, x='species', y='petal length (cm)')
plt.title("Boxplot of Petal Length by Species")
plt.show()
```

```python
sns.boxplot(data=iris_df, x='species', y='petal width (cm)')
plt.title("Boxplot of Petal Width by Species")
plt.show()
```
